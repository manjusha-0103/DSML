```python
from scipy import stats
z = np.abs(stats.zscore(df['maths score']))
#print(z)
#np.abs(stats.zscore(df['maths score']))
z
```

```python
threshold = 0.18
sample_outliers = np.where(z <threshold)
sample_outliers
```

```python
#Detecting outliers using InterQuantile Range(IQR):
#step1:
sorted_score= sorted(df['reading score'])
sorted_score
```

```python
q1 = np.percentile(sorted_score, 25)
q3 = np.percentile(sorted_score, 75)
print(q1,q3)

iqr = q3 - q1
lwr_bound = q1-(1.5*iqr)
upr_bound = q3+(1.5*iqr)
print(lwr_bound,upr_bound)
```

```python
r_outliers = []
for i in sorted_score:
  if (i<lwr_bound or i>upr_bound):
    r_outliers.append(i)
print(r_outliers)
```

```python
# handling Outlier
# Trimming/removing the outlier:
new_df = df
for i in sample_outliers:
    print(i)
    try:
        new_df.drop(i,axis =1)
    except: 
        print("Not found.")

#new_df
#new_df.compare(df).all()
```

```python
#Quantile based flooring and capping:
df_stud = df 
ninetieth_percentile = np.percentile(df_stud['maths score'], 90)
b = np.where(df_stud['maths score']>ninetieth_percentile,ninetieth_percentile, df_stud['maths score'])
print("New array:",b)
```

```python
df_stud.insert(1,"m score",b,True)
df_stud
```
